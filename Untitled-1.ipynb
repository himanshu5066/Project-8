```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer

# ---------------------------
# 2. Preprocess the Data
# ---------------------------

# Create a 'salary' column
def calculate_salary(row):
    if row['is_hourly']:
        # Use average of hourly_low and hourly_high
        try:
            return (float(row['hourly_low']) + float(row['hourly_high'])) / 2
        except:
            return np.nan
    else:
        try:
            return float(row['budget'])
        except:
            return np.nan

df['salary'] = df.apply(calculate_salary, axis=1)

# Drop rows where salary is NaN
df = df.dropna(subset=['salary'])

# Lowercase all titles for consistency
df['title'] = df['title'].str.lower()

# ---------------------------
# 3. Extract Keywords from Titles
# ---------------------------

# Use CountVectorizer to tokenize titles
vectorizer = CountVectorizer(stop_words='english', max_features=100)
X = vectorizer.fit_transform(df['title'])

# Convert to DataFrame
keywords_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Combine with salary
keywords_df['salary'] = df['salary'].values

# ---------------------------
# 4. Analyze Keyword Correlation
# ---------------------------

# Calculate average salary per keyword
keyword_salary = {}
for keyword in keywords_df.columns[:-1]:  # Exclude salary column
    avg_salary = keywords_df.loc[keywords_df[keyword] == 1, 'salary'].mean()
    keyword_salary[keyword] = avg_salary

# Convert to DataFrame for visualization
keyword_salary_df = pd.DataFrame(keyword_salary.items(), columns=['keyword', 'avg_salary'])
keyword_salary_df = keyword_salary_df.dropna()
keyword_salary_df = keyword_salary_df.sort_values(by='avg_salary', ascending=False)

# ---------------------------
# 5. Visualize Top Keywords
# ---------------------------

plt.figure(figsize=(12, 6))
sns.barplot(data=keyword_salary_df.head(15), x='avg_salary', y='keyword', palette='viridis')
plt.title("Top 15 Keywords with Highest Average Salary")
plt.xlabel("Average Salary (USD)")
plt.ylabel("Keyword")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re

# Display basic information
print("Dataset shape:", df.shape)
print(df.head())

# Step 1: Preprocess - Clean and Extract Keywords from Titles
def extract_keywords(title):
    # Remove punctuation, convert to lowercase
    title = re.sub(r'[^\w\s]', '', title.lower())
    return title.split()

df['keywords'] = df['title'].apply(extract_keywords)

# Flatten the keyword list and get frequency
from collections import Counter
all_keywords = [word for keywords in df['keywords'] for word in keywords]
keyword_freq = Counter(all_keywords)

# Convert to DataFrame for plotting
keyword_df = pd.DataFrame(keyword_freq.items(), columns=['keyword', 'frequency'])
keyword_df = keyword_df.sort_values(by='frequency', ascending=False)

# Optional: Remove common stopwords and generic words
stopwords = set([
    'for', 'and', 'the', 'to', 'a', 'in', 'with', 'of', 'on', 'who', 'can',
    'job', 'needed', 'need', 'you', 'looking', 'i', 'we', 'is', 'at', 'my',
    'fix', 'or', 'site'
])
keyword_df = keyword_df[~keyword_df['keyword'].isin(stopwords)]

# Show top 15 emerging keywords
print("\nTop 15 keywords in job titles:")
print(keyword_df.head(15))

# Step 2: Visualize Keyword Frequency
plt.figure(figsize=(12, 6))
sns.barplot(data=keyword_df.head(15), x='frequency', y='keyword', palette='viridis')
plt.title('Top 15 Most Frequent Keywords in Job Titles')
plt.xlabel('Frequency')
plt.ylabel('Keyword')
plt.tight_layout()
plt.show()

# Step 3: Word Cloud of Keywords
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(keyword_freq)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Job Title Keywords')
plt.show()

# Optional: Group similar job categories manually for deeper insights
category_keywords = {
    'developer': ['developer', 'full', 'stack', 'web', 'php', 'html', 'css'],
    'design': ['designer', 'design', 'ui', 'ux', 'logo'],
    'marketing': ['marketing', 'ads', 'affiliate', 'seo'],
    'data': ['data', 'analysis', 'report'],
    'customer_service': ['customer', 'service'],
    'engineering': ['mechanical', 'engineer', '3d'],
    'media': ['media', 'buyer'],
    'shopify': ['shopify'],
    'wordpress': ['wordpress'],
    'bubble': ['bubble'],
    'video': ['video', 'promo'],
    'fashion': ['fashion']
}

# Count how many times each category appears
def categorize(keywords):
    for category, words in category_keywords.items():
        if any(word in keywords for word in words):
            return category
    return 'other'

df['category'] = df['keywords'].apply(categorize)

# Frequency of each category
category_counts = df['category'].value_counts().reset_index()
category_counts.columns = ['category', 'count']

# Visualize category frequency
plt.figure(figsize=(10, 5))
sns.barplot(data=category_counts, x='count', y='category', palette='magma')
plt.title('Job Posting Frequency by Category')
plt.xlabel('Number of Postings')
plt.ylabel('Job Category')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
from sklearn.feature_extraction.text import CountVectorizer

# Convert published_date to datetime
df['published_date'] = pd.to_datetime(df['published_date'], utc=True)

# Extract just the job role from title (simplified approach)
df['job_role'] = df['title'].str.lower().str.extract(r'^(.*?)(?: for| with| needed| who| in| -| |:|,|\.|$)')

# Group by date and job_role
job_trend = df.groupby([df['published_date'].dt.date, 'job_role']).size().reset_index(name='count')

# Pivot for time-series trend analysis
pivot_df = job_trend.pivot(index='published_date', columns='job_role', values='count').fillna(0)

# Select top 5 most frequent roles over time
top_roles = pivot_df.sum().sort_values(ascending=False).head(5).index
pivot_top = pivot_df[top_roles]

# Plot trends
plt.figure(figsize=(14, 6))
pivot_top.plot()
plt.title("Top 5 Job Roles Over Time")
plt.ylabel("Number of Postings")
plt.xlabel("Date")
plt.grid(True)
plt.tight_layout()
plt.show()

# Forecast future demand using Prophet
forecast_results = {}
for role in top_roles:
    role_series = pivot_df[role].reset_index()
    role_series.columns = ['ds', 'y']
    
    model = Prophet()
    model.fit(role_series)
    
    future = model.make_future_dataframe(periods=30)
    forecast = model.predict(future)
    
    forecast_results[role] = (model, forecast)
    
    # Plot forecast
    model.plot(forecast)
    plt.title(f"Forecast for Job Role: {role.title()}")
    plt.show()

# Display job roles with the highest predicted growth
growth_rates = {
    role: forecast_results[role][1].iloc[-30:]['yhat'].mean() for role in top_roles
}
print("Predicted Average Postings in Next 30 Days:")
print(pd.Series(growth_rates).sort_values(ascending=False))
```

```python
import pandas as pd

# Convert empty strings to NaN
df = df.replace("", pd.NA)

# Count total cells
total_cells = df.size

# Count missing cells
missing_cells = df.isna().sum().sum()

# Completeness Accuracy (% of non-missing data)
completeness_accuracy = 100 * (total_cells - missing_cells) / total_cells

# Logical consistency checks
# 1. hourly_low <= hourly_high if is_hourly is True
df_hourly = df[df['is_hourly'] == True].copy()
df_hourly[['hourly_low', 'hourly_high']] = df_hourly[['hourly_low', 'hourly_high']].apply(pd.to_numeric, errors='coerce')
consistent_hourly = df_hourly[df_hourly['hourly_low'] <= df_hourly['hourly_high']]

# 2. If is_hourly is False, then budget should be present
df_fixed = df[df['is_hourly'] == False].copy()
df_fixed_budget = df_fixed[df_fixed['budget'].notna()]

# Total logical consistency rows
total_consistent = len(consistent_hourly) + len(df_fixed_budget)
total_relevant = len(df_hourly) + len(df_fixed)
logical_accuracy = 100 * total_consistent / total_relevant if total_relevant > 0 else 0

# Final report
print(f"Completeness Accuracy: {completeness_accuracy:.2f}%")
print(f"Logical Accuracy: {logical_accuracy:.2f}%")
print(f"Overall Accuracy Estimate: {(completeness_accuracy + logical_accuracy) / 2:.2f}%")
```
